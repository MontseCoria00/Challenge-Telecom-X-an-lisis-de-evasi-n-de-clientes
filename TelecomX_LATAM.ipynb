### Limpieza y Tratamiento de Datos

El proceso de limpieza y tratamiento de datos fue fundamental para preparar el conjunto de datos de Telecom X para el análisis. Los pasos clave incluyeron:

1.  **Importación de Datos:** Los datos se cargaron inicialmente desde un archivo JSON (`TelecomX_Data.json`) utilizando la librería `pandas` para crear un DataFrame llamado `datos`.

2.  **Normalización de Columnas Anidadas:** El DataFrame original contenía columnas con estructuras de diccionario (`customer`, `phone`, `internet`, `account`). Para extraer las características individuales, se aplicó `pd.json_normalize()` a cada una de estas columnas. Luego, las columnas originales anidadas se eliminaron y las nuevas características se concatenaron al DataFrame principal. Este proceso se realizó en las siguientes etapas:
    *   **Columna 'account':** Se normalizó para extraer `Contract`, `PaperlessBilling`, `PaymentMethod`, `Charges.Monthly` y `Charges.Total`.
    *   **Columna 'customer':** Se normalizó para obtener `gender`, `SeniorCitizen`, `Partner`, `Dependents` y `tenure`.
    *   **Columna 'phone':** Se normalizó para `PhoneService` y `MultipleLines`.
    *   **Columna 'internet':** Se normalizó para `InternetService`, `OnlineSecurity`, `OnlineBackup`, `DeviceProtection`, `TechSupport`, `StreamingTV` y `StreamingMovies`.

3.  **Manejo Inicial de Valores Vacíos en 'Churn':** La columna `Churn` presentaba valores vacíos (`''`). Estos se reemplazaron por 'No', asumiendo que un valor ausente en esta columna indica que el cliente no ha cancelado el servicio.

4.  **Conversión de Categorías Binarias a Numéricas:** Muchas columnas categóricas que contenían respuestas 'Yes'/'No' (o variantes como 'Female'/'Male', 'No phone service', 'No internet service') se mapearon a valores binarios (1 y 0) para facilitar el análisis numérico. Esto se aplicó a `Churn`, `PaperlessBilling`, `Partner`, `Dependents`, `PhoneService`, `MultipleLines`, `OnlineSecurity`, `OnlineBackup`, `DeviceProtection`, `TechSupport`, `StreamingTV`, `StreamingMovies` y `gender`.

5.  **Manejo de 'Charges.Total':** Esta columna era inicialmente de tipo 'object' y podía contener cadenas vacías. Se convirtió a tipo numérico (`float64`), y cualquier valor no numérico o cadena vacía se reemplazó por `NaN` y posteriormente se rellenó con `0`.

6.  **One-Hot Encoding para Columnas Categóricas Restantes:** Para las columnas categóricas con más de dos valores únicos que no eran binarias (`InternetService`, `PaymentMethod`, `Contract`), se aplicó One-Hot Encoding utilizando `pd.get_dummies()`. Esto creó nuevas columnas binarias para cada categoría, y se utilizó `drop_first=True` para evitar la multicolinealidad.

7.  **Relleno de Valores Nulos Final:** Tras todas las transformaciones y codificaciones, se identificaron y rellenaron los valores nulos restantes en columnas como `Churn`, `MultipleLines`, `OnlineSecurity`, `OnlineBackup`, `DeviceProtection`, `TechSupport`, `StreamingTV` y `StreamingMovies` con `0`. Esto se hizo bajo la suposición de que un valor nulo en estas columnas binarias indicaba la ausencia del servicio o característica.

8.  **Eliminación de la Columna 'customerID':** La columna `customerID` se eliminó del DataFrame final ya que es un identificador único y no aporta valor predictivo al modelo.

 Conclusión

Tras un exhaustivo análisis de los datos de Telecom X, se han extraído las siguientes conclusiones que se consideran clave respecto a la evasión de clientes (Churn):

Impacto de la Antigüedad del Cliente : La tasa de cancelacón es significativamente alta dentro de los primeros meses de servicio. Se observa un pico importante de abandono en el primer mes, que disminuye gradualmente a medida que el cliente permanece más tiempo con la compañía. Esto sugiere que los clientes nuevos son más propensos a irse, posiblemente se deba a expectativas no cumplidas, problemas de adaptación o la búsqueda de mejores ofertas iniciales.

Métodos de Pago y cancelación: Los clientes que utilizan "Cheques Electrónicos" presentan una tasa de Churn considerablemente más alta en comparación con otros métodos de pago. Esto podría indicar insatisfacción con el proceso de pago, problemas de seguridad percibidos, o que este método es preferido por clientes menos comprometidos a largo plazo. Por otro lado, los métodos automáticos (tarjetas de crédito) muestran tasas de cancelaci;on más bajas.

Tipo de Contrato: Los clientes con contratos "Mes a Mes" tienen una tasa de cancelación mucho mayor que aquellos con contratos a "Uno" o "Dos Años". Esto es esperado, ya que los contratos más cortos ofrecen mayor flexibilidad para cambiar de proveedor, pero resalta la necesidad de estrategias de retención específicas para este segmento. Los contratos a largo plazo demuestran una mayor lealtad.

Cargos Mensuales y cancelaacón: Se observa que los clientes con "Cargos Mensuales" más altos tienden a tener una mayor probabilidad de Churn. Esto podría deberse a una percepción de que no están obteniendo suficiente valor por el precio, o que la competencia ofrece servicios similares a costos más bajos.

Cargos Totales y Churn: A diferencia de los cargos mensuales, los clientes con "Cargos Totales" más bajos muestran una mayor propensión a la cancelación. Esto refuerza la idea de que los clientes nuevos (con cargos totales bajos) son más susceptibles a irse. A medida que el cargo total aumenta (lo que implica mayor antigüedad), la probabilidad de baja disminuye.
